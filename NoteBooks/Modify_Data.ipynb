```python
def get_all(y,m):
    indx_date = []
    days = monthrange(y,m)[1]
    start_date = datetime.date(y, m, 1)
    end_date = datetime.date(y, m, days)
    delta = datetime.timedelta(days=1)
    while start_date <= end_date:
        indx_date.append(start_date)
        start_date += delta
    return indx_date
```

```python
Row_list =[] 
date = []
for i in range((df.shape[0])):
    #ll = monthrange
    #Row_list.append(list(df.iloc[i, 0:2]))
    yy = df.YEAR[i]
    mm = df.MN[i]
    #print(yy)
    ll = monthrange(yy,mm)[1]
    Row_list.append(list(df.iloc[i, 2:ll+2]))
    date.append(get_all(yy,mm))
#Print the list 
print(Row_list) 
```
